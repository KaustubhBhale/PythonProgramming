```python
from random import *

class Dice:
    def __init__(self,sides):
        self.sides = sides
        
    def roll_dice(self):
        return randint(1,self.sides)

dice1 = Dice(6)
print(dice1.roll_dice())
```

```python
import math

class Circle:
    def __init__(self,radius):
        self.radius=radius
        
    def area(self):
        return math.pi*self.radius*self.radius
    def perimeter(self):
        return 2*math.pi*self.radius


c1 = Circle(7)
print('Area is - ',c1.area())
print('Permieter is - 'c1.perimeter())
```

```python
#Instance and class variable  -> 

class Employee:
    employee_count=100
    def __init__(self,name,salary,designation):
        self.name=name
        self.salary=salary
        self.designation=designation
        self.empid='e'+ str(Employee.employee_count)
        Employee.employee_count+=1
    def show_details(self):
        print('Employee Id : ',self.empid)
        print('Employee Name : ',self.name)
        print('Employee Salary : ',self.salary)
        print('Employee Designation : ',self.designation)
    @classmethod    
    def total_employees(self):
        return Employee.employee_count-100
        

emp1 = Employee('Rajesh',50000,'Manager')
emp2 = Employee('Mahesh',70000,'Sr Manager')
print('Employee1 details - ',emp1.show_details())
print('----------------------------------------')
print('Employee2 details - ',emp2.show_details())
print('----------------------------------------')
print('Total employees is ',emp2.total_employees())
```

```python
#Static Methods  -> There can be some functionality that relates to the class, but does not require any instance(s) to do some work, static methods can be used in such cases. A static method is a method which is bound to the class and not the object of the class. 
class Calculator:
    @staticmethod
    def add(a,b):
        return a+b
    @staticmethod
    def sub(a,b):
        return a-b

print(Calculator.add(2,3))
print(Calculator.sub(10,5))    
```

```python
#Accessors and Mutators
class Customer:
    def __init__(self,name,phone):
        self.name=name
        self.phone=phone
    def getName(self):
        return self.name
    def getPhone(self):
        return self.phone
    def setPhone(self,newPhone):
        self.phone=newPhone

c1 = Customer('Amit','9856689876')
print('Original - ',c1.getName(),c1.getPhone())
c1.setPhone('00000000000')
print('Updated details are - ',c1.getName(),c1.getPhone())
```

```python
#Currency converter 
class CurrencyConverter:
    def __init__(self,currency,rate):
        self.currency=currency
        self.rate=rate
    def setCurrency(self,currency):
        self.currency=currency
    def getCurrency(self):
        return self.currency
    def setRate(self,rate):
        self.rate=rate
    def getRate(self):
        return self.rate
    def convert(self,amount):
        return self.currency + ' conversion is '+str(amount*self.rate)

cc = CurrencyConverter('USD',70)
print('Currency : ',cc.getCurrency())
print('Current Conversion rate ',cc.getRate())
print(cc.convert(70))

aud = CurrencyConverter('AUD',65)
aud.setCurrency('AUD')
aud.setRate(65)

print('--------------------------------------------')

print('Currency : ',aud.getCurrency())
print('Current Conversion rate ',aud.getRate())
print(aud.convert(70))
```

```python
#Inheritance
import math

class Polygon:
    def __init__(self,ns,*sides):
        self.no_of_sides=ns
        self.sides = sides
class Triangle(Polygon):
    def __init__(self,ns,*sides):
        super().__init__(ns,*sides)
    def area(self):
        a,b,c=self.sides
        s=(a+b+c)/2
        a=math.sqrt(s*(s-a)*(s-b)*(s-c))
        return a
t1 = Triangle(3,10,15,9)
print(t1.area())
    
```

```python
#Inner and Outer Classes -> printing course details along with suggested books. Book is a inner class to the Course Outer class

#example of variable number of arguments
class Course:
    def __init__(self,courseName,courseDuration,*suggestedBooks):
        self.courseName=courseName
        self.courseDuration=courseDuration
        self.suggestedBooks = [self.Book(b) for b in suggestedBooks]
    def course_details(self):
        print('Course Name: ',self.courseName)
        print('Course Duration(hrs): ',self.courseDuration)
        print('Suggested Books -  ')
        for b in self.suggestedBooks:
            print(b.__str__())
    class Book:
        def __init__(self,title):
            self.title=title
        def __str__(self):
            return self.title

c1 = Course('Python',10,'Python by Shah','Python for Data Science','Python for AI and ML')
c1.course_details()
```

```python
#Inner class

class Computer:
    def __init__(self,computerName,make,os):
        self.computerName=computerName
        self.make=self.CPU(make)
        self.os=self.OS(os)
    class CPU:
        def __init__(self,make):
            self.make=make
        def getMake(self):
            return self.make
    class OS:
        def __init__(self,OSname):
            self.OSname=OSname
        def getOSName(self):
            return self.OSname
    def __str__(self):
        return 'Name : ',self.computerName,'\nCPU : ',self.make.getMake(),'\ nOS Name: ',self.os.getOSName()

PC1 =Computer('PC1','Intel','Windows')
print(PC1.__str__())
```

```python
#Demonstration of Polymorphism -> same function name (but different signatures) being used for different types. 
class Dog:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def sound(self):
        return "Bhow Bhow"
    def __str__(self):
        return 'Name: ',self.name,'Age: ',self.age
class Cat:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def sound(self):
        return "Meow Meow"
    def __str__(self):
        return 'Name: ',self.name,'Age: ',self.age

def myPet(pet):
    print(pet.__str__())
    print(pet.sound())
    

dog = Dog('Moti',3)
cat = Cat('Rani',6)

myPet(dog)
print()
myPet(cat)
```
